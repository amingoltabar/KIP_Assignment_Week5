```python
# Step 2: Missing Values
# Step 2.1
# Function to perform custom fill and delete for the 'case' column
def custom_fill_and_delete(df):
    i = 1  # Start with the second row
    while i < len(df) - 1:
        if pd.isna(df.at[i, 'case']):
            prev_value = df.at[i - 1, 'case']
            next_value = df.at[i + 1, 'case']
            if prev_value == next_value:
                df.at[i, 'case'] = prev_value
                i += 1  # Move to the next row after processing
            else:
                df.drop(index=i, inplace=True)
                # Reset the index to clean up the index after rows are deleted
                df.reset_index(drop=True, inplace=True)
        else:
            i += 1  # Move to the next row if no missing value
    return df

# Apply the custom fill and delete function to the 'case' column
my_data = custom_fill_and_delete(my_data)
```

```python
# Step 2: Missing Values
# Step 2.3: # Fill missing values in categorial column with the most frequent value for their respective cases

# Define a function to fill missing values in categorial columns with the most frequent value for their respective cases
def fill_missing_with_mode_by_case(data, case_column, categorical_columns):
    for cat_col in categorical_columns:
        grouped = data.groupby(case_column)  # Step 1: Group the dataset by the 'case' column
        case_modes = grouped[cat_col].apply(lambda x: x.mode().iloc[0])  # Step 2: Calculate the most frequent value (mode) of the attribute for each case
        
        # Step 3: Fill missing values in 'SIRSCriteria2OrMore' with the most frequent value for their respective cases
        for case, mode_value in case_modes.items():
            my_data.loc[(my_data[case_column] == case) & my_data[cat_col].isna(), cat_col] = mode_value
    return data
```

```python
# Each case will be aggregated into one row, with the most frequent value for each variable.

# Step 1: Group the dataset by the 'case' column
grouped = my_data.groupby('case')

# Step 2: Define a function to calculate the most frequent value for a column within a group
def most_frequent(column):
    return column.value_counts().idxmax()

# Step 3: Apply the most_frequent function to each column for each case
# summary_my_data will contain a single row for each case, with the most frequent values for each column.
summary_data = grouped.agg({
    'CRP': most_frequent,
    'Leucocytes': most_frequent,
    'LacticAcid': most_frequent,
    'Age': most_frequent,
    'SIRSCriteria2': most_frequent,
    'SIRSCriteria3': most_frequent,
    'SIRSCriteria4': most_frequent,
    'duration': 'first'  # Select the first value of duration, assuming it's the same for each case
}).reset_index()
```

```python
# Question b
# Perform one-hot encoding for these four columns: 'CRP', 'Leucocytes', 'LacticAcid', 'Age'
my_data = pd.get_dummies(my_data, columns=['CRP', 'Leucocytes', 'LacticAcid', 'Age','duration'])
```

```python
# Find frequent itemsets
frequent_itemsets = apriori(my_data, min_support=0.1, use_colnames=True)
```

```python
frequent_itemsets
```

```python
# Generate association rules with the minimum confidence of 0.2
my_association_rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.2)

# Filter the rules with a maximum of 3 items in the antecedent
filtered_rules = my_association_rules[(my_association_rules['antecedents'].apply(len) <= 3) & (my_association_rules['consequents'].apply(len) == 1)]
```

```python
filtered_rules
```

```python
# Filter further to keep only rules where the consequent is one of 'duration' items
important_rules = filtered_rules[filtered_rules['consequents'].apply(lambda x: list(x)[0] in 
                                                                     ['duration_Min – Q1','duration_Q1 - Q2',  'duration_Q2 – Q3', 'duration_Q3 – max'])]
```

```python
important_rules
```

```python
# Filter association rules based on minimum lift
final_rules = important_rules[my_association_rules['lift'] > 0.9]
```

```python
# final_rules is sorted based on the lift in descending order (highest lift first)
final_rules = final_rules.sort_values(by='lift', ascending=False)
final_rules
```
